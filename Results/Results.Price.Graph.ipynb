```python
def load_data(filename):
    dataset_total = pd.read_csv(filename,na_values='n/a',parse_dates=['Date'])

# Set index
    dataset_total.set_index('Date', inplace=True)
    dataset_total= dataset_total.loc[:, (dataset_total!= 0).any(axis=0)]
    return dataset_total
```

```python
dataset_total=load_data("ZEC-USD.csv")
dataset_total.head()
```

```python
fig_1 = plt.figure(figsize=(10, 7))
ax= fig_1.add_subplot(111)
ax.set_title('Zcash Price in US Dollars')
plt.plot(dataset_total.loc['2016-10-27':'2016-11-30','Adj Close'], color = 'black', label = 'Zcash',linewidth=2)
plt.xlabel('Time') 
plt.ylabel('Dollars')
plt.legend()
#plt.style.use('ggplot')
plt.show()
```
